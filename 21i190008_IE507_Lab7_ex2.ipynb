```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cars=pd.read_csv('cars.csv')
```


### 2)

```python
sns.histplot(cars.mpg,kde=True)
```

```python
for j in [20,50,100,200,500]:
    print('Number of bins:',j)
    sns.histplot(cars.mpg,bins=j,kde=True )
    plt.show()
    print('\n'*2)
```

The scale of the kde curve keeps decreasing as the number of bins increases, although its shape is more or less the same. This is because each bin gets divided into bins of smaller sizes and the number of observations would be divided among these smaller bins. 

### 3)

```python
plt.figure(figsize=(16,8))
plot3=sns.barplot(cars.mpg,cars.displacement)
sort = sorted(plot3.patches,reverse=True,key=lambda i:i.get_height())
x1 = sort[0].get_x()
y1 = sort[0].get_height()
x2 = sort[1].get_x()
y2 = sort[1].get_height()
plot3.set_xticklabels(plot3.get_xticklabels(),rotation=90,fontsize=7)
plt.annotate('Highest mean! '+str(y1),xy=(x1,y1),xytext=(x1+0.3,y1+40.5),arrowprops={'facecolor':'green','arrowstyle':'->'})
plt.annotate('2nd highest mean! '+str(y2),xy=(x2,y2),xytext=(x2+10,y2+20.5),arrowprops={'facecolor':'red','arrowstyle':'->'})
```

From the graph, we see that, as MPG increases, average displacement of the vehicles decrease, according to an exponential curve. 

### 4)

```python
plt.figure(figsize=(20,8))
plot4=sns.barplot(x='mpg',y='displacement',data=cars,hue='model_year',estimator=np.median,ci=None)
n = [i for i in plot4.patches if not np.isnan(i.get_height())]
sort2=sorted(n,reverse=True,key = lambda i: i.get_height())
x21 = sort2[0].get_x()
y21 = sort2[0].get_height()
x22 = sort2[1].get_x()
y22 = sort2[1].get_height()
plot4.set_xticklabels(plot4.get_xticklabels(),rotation=90,fontsize=7)
plt.annotate(str(sort2[0].get_height()),xy=(x21,y21),xytext=(x21+3.3,y21+5),arrowprops={'arrowstyle':'->'})
plt.annotate(str(sort2[1].get_height()),xy=(x22,y22),xytext=(x22+2,y22+50.5),arrowprops={'arrowstyle':'->'})

plot4.legend(loc='upper right',ncol=2)
```

For a given mpg, the median displacement over all years is more or less the same, with the exception of the year 1979, which has a comparatively higher displacemnt for most mpg values.

### 5)

```python
plt.figure(figsize=(16,8))
plot5=sns.barplot(x='mpg',y='displacement',data=cars,hue='origin',estimator=np.median,ci=None)
m = [i for i in plot5.patches if not np.isnan(i.get_height())]
sort5=sorted(m,key = lambda i: i.get_height())

x51 = sort5[0].get_x()
y51 = sort5[0].get_height()
x52 = sort5[1].get_x()
y52 = sort5[1].get_height()

plt.annotate(str(sort5[0].get_height()),xy=(x51+0.05,y51),xytext=(x51+2.3,y51+250.5),arrowprops={'arrowstyle':'->'})
plt.annotate(str(sort5[1].get_height()),xy=(x52+0.05,y52),xytext=(x52+3.3,y52+150.5),arrowprops={'arrowstyle':'->'})

plot5.set_xticklabels(plot5.get_xticklabels(),rotation=90,fontsize=7)
plt.show()
```

From the graph , we can see that Japan doesn't produce cars with mpg value less than 18. USA doesn't produce cars with mpg above 39. Europe doesn't produce cars with mpg lower than 16.2.\
For a given mpg, cars produced in USA have a higher median displacement than the other 2 regions.

### 8)

```python
sns.lineplot(x='model_year',y='horsepower',data=cars)
```

We can see that there is a weak negative correlation between horsepower and model. As the years go by, the average horsepower of the produced cars decrease, with the exception of the spell from 1971 - 1973 where there was an increase.

### 9)

```python
plot9=sns.boxplot(x='model_year',y='weight',data=cars)
c70 = 0,np.percentile(cars.weight[cars.model_year==70],75)
c74 = 4,np.percentile(cars.weight[cars.model_year==74],75)
plot9.annotate(c70[1],xy = c70,xytext=(-0.1,4800),arrowprops={'arrowstyle':'->'},fontsize=8 )
plot9.annotate(c74[1],xy = c74,xytext=(4,4800),arrowprops={'arrowstyle':'->'},fontsize=10)
```

There is no monotonic relation between median weights and model year, but , roughly speaking, we can say that the max weight of the cars is decreasing as years go by. The minimum weight has stayed more or less constant

### 10)

```python
plt.figure(figsize=(16,8))
sns.boxplot(x='model_year',y='displacement',data=cars,hue='origin')
```

We can see that the median displacement in cars produced in USA are much higher than other regions, in all the years recorded. However, over the years, the magnitude of the difference has been decreasing. for the other regions, the median displacement has stayed more or less constant over the years (around 100 units).

### 11)

```python
plt.figure(figsize=(15,8))
sns.violinplot(x='model_year',y='acceleration',data=cars,hue='origin')
```

On the whole, the spread of the accelaration is more in the cars from Europe, thus we can say that they offer a wider range of accelarations in their cars. the violins corresponding to the regions USA and Japan are much wider in the middle than those of europe, Thus, we can say that USA and Japan produce their vehicles with accelaration in a tighter range.
